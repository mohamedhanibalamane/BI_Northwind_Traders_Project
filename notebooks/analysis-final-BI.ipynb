```python
# analysis-final-BI.ipynb
# Notebook d'analyse des donn√©es Northwind
# Version FINAL BI
# Affiche les images *_final-BI.png et cr√©e fichiers HTML individuels et global

# ------------------------------
# Import des biblioth√®ques
# ------------------------------
import pandas as pd
from IPython.display import Image, display
import os

# ------------------------------
# Cr√©ation des chemins
# ------------------------------
# On suppose que 'figures' existe d√©j√†
figures_path = "../figures"  # <-- chemin relatif depuis 'notebooks' vers le dossier figures

if not os.path.exists(figures_path):
    raise FileNotFoundError(f"Le dossier '{figures_path}' n'existe pas. V√©rifie le chemin !")

# ------------------------------
# Section 1 : Chargement des images g√©n√©r√©es
# ------------------------------
image_files = [f for f in os.listdir(figures_path) if f.endswith("_final-BI.png")]

print(f"{len(image_files)} images *_final-BI.png trouv√©es dans '{figures_path}':")
for img in image_files:
    print(f" - {img}")

# ------------------------------
# Section 2 : Affichage des images dans le notebook
# ------------------------------
print("\nAffichage des images dans le notebook :\n")
for img in image_files:
    display(Image(os.path.join(figures_path, img)))

# ------------------------------
# Section 3 : Cr√©ation des fichiers HTML individuels
# ------------------------------
html_files = []
for img_file in image_files:
    html_file = os.path.join(figures_path, img_file.replace(".png", ".html"))
    html_content = f"""
    <html>
    <head><title>{img_file}</title></head>
    <body>
    <h2>{img_file}</h2>
    <img src="{img_file}" alt="{img_file}" style="max-width:90%;">
    </body>
    </html>
    """
    with open(html_file, "w", encoding="utf-8") as f:
        f.write(html_content)
    html_files.append(html_file)

print("\n‚úÖ Fichiers HTML individuels cr√©√©s :")
for f in html_files:
    print(f" - {f}")

# ------------------------------
# Section 4 : Cr√©ation du fichier HTML global (dashboard total)
# ------------------------------
dashboard_html = os.path.join(figures_path, "dashboard_total_final-BI.html")
with open(dashboard_html, "w", encoding="utf-8") as f:
    f.write("<html><head><title>Dashboard Northwind FINAL BI</title></head><body>\n")
    f.write("<h1>Dashboard Northwind ‚Äì Version FINAL BI</h1>\n")
    for img_file in image_files:
        f.write(f"<h3>{img_file}</h3>\n")
        f.write(f'<img src="{img_file}" alt="{img_file}" style="max-width:90%;"><br><br>\n')
    f.write("</body></html>")

print(f"\n‚úÖ Fichier HTML global cr√©√© : {dashboard_html}")

# ------------------------------
# Section 5 : Analyse et suivi simple
# ------------------------------
print("\n--- Analyse et suivi des images ---")
print(f"Nombre total d'images affich√©es : {len(image_files)}")
print(f"Dossier des images et fichiers HTML : {figures_path}")
```
